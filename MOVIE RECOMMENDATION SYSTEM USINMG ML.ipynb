# Importing the dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

```

# Data Collection and Pre-processing

```python
# loading the data from the csv file to apandas dataframe
movies_data = pd.read_csv('movies.csv')
```

```python
movies_data.head()
```

```python
# number of rows and columns in the data frame
movies_data.shape
```

```python
# selecting the relevant features for recommendation

selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
# replacing the null values with null string

for feature in  selected_features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining all the 5 selected features

combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']+' '
```

```python
print(combined_features)
```

```python
# converting the text data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

# Cosine Similarity

```python
# getting the similarity scores using cosine similarity
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
# getting the movie name from the user

movie_name = input(' Enter your favourite movie name : ')
```

```python
# creating a list with all the movie names given in the dataset

list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
# finding the close match for the movies name given by the user 

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with title

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the inbox

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
    inbox = movie[0]
    title_from_inbox = movies_data[movies_data.index==inbox]['title'].values[0]
    if (i<30):
        print(i, '.',title_from_inbox)
        i+=1
```

# Movie Recommendation System

```python
# Get movie index from title
def get_title_from_index(index):
    return movies_data[movies_data.index == index]['title'].values[0]

def get_index_from_title(title):
    return movies_data[movies_data.title.str.lower() == title.lower()].index.values[0]

# Input from user
movie_user_likes = input("Enter your favourite movie name : ")
try:
    movie_index = get_index_from_title(movie_user_likes)
except IndexError:
    print("Movie not found in database.")
    exit()

# Get similarity scores
similar_movies = list(enumerate(similarity[movie_index]))  # similarity is your cosine similarity matrix

# Sort similar movies
sorted_similar_movies = sorted(similar_movies, key=lambda x: x[1], reverse=True)

print("\nMovies suggested for you:\n")
i = 1
for movie in sorted_similar_movies:
    index = movie[0]
    if index != movie_index:  # skip the same movie
        print(i, '.', get_title_from_index(index))
        i += 1
        if i > 30:
            break
```
